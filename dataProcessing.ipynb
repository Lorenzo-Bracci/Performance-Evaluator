```python
import numpy as np
import csv
import matplotlib.pyplot as plt

dropRate = list()
fps = list()
bps = list()
filenames = ['res1.csv', 'res3.csv', 'res6.csv', 'res9.csv', 'res12.csv']

for filename in filenames:
    with open(filename) as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        currentDrop = list()
        currentFps = list()
        currentBps = list()
        for row in reader:
            currentDrop.append(float(row[0]))
            currentFps.append(float(row[1]))
            currentBps.append(float(row[2]))
        dropRate.append(currentDrop)
        fps.append(currentFps)
        bps.append(currentBps)

plt.boxplot(bps)
```

```python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

latencyFile = 'lres1.csv'
latency = list()
with open(latencyFile) as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        for row in reader:
            latency.append(np.asfarray(row[:-1],float))
flierprops = dict(marker='x', markerfacecolor='r', markersize=2,
                  linestyle='none', markeredgecolor='r')

plt.boxplot(latency[0:50], flierprops=flierprops)
plt.xticks([10, 20, 30 , 40, 50], [10, 20, 30 , 40, 50])
plt.xlabel("Frame Number")
plt.ylabel("Time in milliseconds")
plt.savefig('latency1', dpi=600)
```

```python
latencyFilenames = ['lres1.csv', 'lres3.csv', 'lres6.csv', 'lres9.csv', 'lres12.csv']
latency = list()

for latencyFile in latencyFilenames:
    with open(latencyFile) as csvfile:
            reader = csv.reader(csvfile, delimiter=',')
            currentLatency = list()
            for row in reader:
                currentLatency.append(np.asfarray(row[:-1],float))
            latency.append([item for sublist in currentLatency for item in sublist])  
            
flierprops = dict(marker='x', markerfacecolor='r', markersize=2,
                  linestyle='none', markeredgecolor='r')

plt.boxplot(latency, flierprops=flierprops)
plt.xticks([1, 2, 3, 4, 5], [1, 3, 6, 9, 12])
plt.xlabel("Number of Threads used")
plt.ylabel("Time in milliseconds")
plt.savefig('latency1', dpi=600)
```

```python
bufferFile = "pres12.csv"
timeToFill = list()

with open(bufferFile) as csvfile:
            reader = csv.reader(csvfile, delimiter=',')
            current50 = list()
            current80 = list()
            current95 = list()
            current99 = list()
            current100 = list()
            for row in reader:
                current50.append(float(row[0]))
                current80.append(float(row[1]))
                current95.append(float(row[2]))
                current99.append(float(row[3]))
                current100.append(float(row[4]))
            timeToFill.append(current50)
            timeToFill.append(current80)
            timeToFill.append(current95)
            timeToFill.append(current99)
            timeToFill.append(current100)

flierprops = dict(marker='x', markerfacecolor='r', markersize=2,
                  linestyle='none', markeredgecolor='r')
plt.boxplot(timeToFill, positions=[50,80,95, 99, 100], flierprops=flierprops)
#plt.xticks([1, 2, 3, 4, 5], [50, 80, 95, 99, 100])
plt.xlabel("Elements filled in buffer")
plt.ylabel("Time in milliseconds")
plt.savefig('latency1', dpi=600)           
```
